```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
headers = {
    'User-Agent': 'Chrome/92.0.4515.159'
}
```

```python
URL_cs = 'https://cfd.nu.edu.pk/department-cs/'
URL_se = 'https://cfd.nu.edu.pk/department-se/'
URL_ai = 'https://cfd.nu.edu.pk/department-ai/'
URL_ee = 'https://cfd.nu.edu.pk/department-ee/'
URL_sh = 'https://cfd.nu.edu.pk/department-sh/'
URL_fsm = 'https://cfd.nu.edu.pk/department-fsm/'
cs = requests.get(URL_cs, headers=headers)
print(cs)
se = requests.get(URL_se, headers=headers)
print(se)
ai = requests.get(URL_ai, headers=headers)
print(ai)
ee = requests.get(URL_ee, headers=headers)
print(ee)
sh = requests.get(URL_sh, headers=headers)
print(sh)
fsm = requests.get(URL_fsm, headers=headers)
print(fsm)
URL_department = 'https://cfd.nu.edu.pk/all-departments/'
department = requests.get(URL_department, headers=headers)
print(department)
```

```python
soup0 = BeautifulSoup(cs.content, 'html.parser')
soup1 = BeautifulSoup(se.content, 'html.parser')
soup2 = BeautifulSoup(se.content, 'html.parser')
soup3 = BeautifulSoup(ee.content, 'html.parser')
soup4 = BeautifulSoup(sh.content, 'html.parser')
soup5 = BeautifulSoup(fsm.content, 'html.parser')
soup6 = BeautifulSoup(department.content, 'html.parser')
```

```python
department_name = []
department_name.append(soup6.find('li', id='menu-item-508095').text.split('\n\n')[1].split('Department of ')[1].strip())
department_name.append(soup6.find('li', id='menu-item-508095').text.split('\n\n')[3].split('\n')[1].split('Department of ')[1].strip())
department_name.append(soup6.find('li', id='menu-item-508095').text.split('\n\n')[5].split('\n')[1].split('Department of ')[1].strip())
department_name.append(soup6.find('li', id='menu-item-5598').text.split('\n\n')[0].split('Department of ')[1].strip())
department_name.append(soup6.find('li', id='menu-item-5597').text.split('\n\n')[0].split('FAST School of ')[1].strip())
department_name.append(soup6.find('li', id='menu-item-5600').text.split('\n\n')[0].split('Department of ')[1].strip())
print(department_name)

school_name = []
school_name.append(soup6.find('li', id='menu-item-508095').text.split('\n\n')[0])
school_name.append(soup6.find('li', id='menu-item-508095').text.split('\n\n')[0])
school_name.append(soup6.find('li', id='menu-item-508095').text.split('\n\n')[0])
school_name.append(soup6.find('li', id='menu-item-5598').text.split('\n\n')[0])
school_name.append(soup6.find('li', id='menu-item-5597').text.split('\n\n')[0])
school_name.append(soup6.find('li', id='menu-item-5600').text.split('\n\n')[0])
print(school_name)
```

```python
faculty = []
faculty.append(soup0.findAll('div', class_='unitech-teacher'))
faculty.append(soup1.findAll('div', class_='unitech-teacher'))
faculty.append(soup2.findAll('div', class_='unitech-teacher'))
faculty.append(soup3.findAll('div', class_='unitech-teacher'))
faculty.append(soup4.findAll('div', class_='unitech-teacher'))
faculty.append(soup5.findAll('div', class_='unitech-teacher'))
```

```python
# Create the DataFrame
faculty_chiniot_data = pd.DataFrame(columns=['ID',
                                     'Name',
                                     'Designation',
                                     'HEC Approved PHD Supervisor',
                                     'Highest Education',
                                     'Email',
                                     'School',
                                     'Department',
                                     'Extension',
                                     'ImageURL'])
```

```python
faculty_chiniot_data.drop(faculty_chiniot_data.index, inplace=True)
count = 0
for faculty_array in faculty:
    depertment = department_name[count]
    school = school_name[count]
    count += 1
    for professor in faculty_array:
        id = 'Not Available'
        print('ID: ', id)
        
        name = str(professor.find('div', class_='unitech-teacher__details').find('div', class_='unitech-teacher__inner').find('h4').text.strip())
        print('Name: ', name)
        
        designation = str(professor.find('div', class_='unitech-teacher__details').find('div', class_='unitech-teacher__inner').find('h6').text.strip())
        print('Designation: ', designation)
        
        info = professor.find('div', class_='unitech-teacher__details').find('ul', class_='unitech-teacher-social-icon').findAll('p')
        
        if len(info) > 1:
            hec = True
            print('HEC: ', hec)
        else:
            hec = False
            print('HEC: ', hec)
        
        email = str(info[0].text.strip())
        if email:
            print('Email: ', email)
        else:
            email = 'Not Available'
            print('Email: ', email)
        
        school_str = str(school)
        print('School: ', school_str)
        
        depertment_str = str(depertment)
        print('Department: ', depertment_str)
        
        img = professor.find('div', class_='unitech-teacher__thumb').find('img').get('src')
        print('IMG: ', img)
        
        profile_url = professor.find('div', class_='unitech-teacher__details').find('div', class_='unitech-teacher__inner').find('h4').find('a').get('href')
        
        p_url = requests.get(profile_url, headers=headers)
        soup6 = BeautifulSoup(p_url.content, 'html.parser')
        
        if p_url.status_code == 200:
            edu_array = soup6.find('div', class_='htc__skill__container progress__bar--2')
        
            if edu_array:
                edu_array_ul = edu_array.find('ul')
        
                if edu_array_ul:
                    edu_array_lis = edu_array_ul.find_all('li')
        
                    if len(edu_array_lis) > 0:
                        edu = str(edu_array_lis[0].text.strip())
                        if edu:
                            print('Education: ', edu)
                        else:
                            if len(edu_array_lis) > 1:
                                edu = str(edu_array_lis[1].text.strip())
                                print('Education: ', edu)
                            else:
                                edu = 'Not Available'
                                print('Education: ', edu)
                    else:
                        edu = 'Not Available'
                        print('Education: ', edu)
        
                elif edu_array.find('ol'):
                    edu_array_ol = edu_array.find('ol')
                    edu_array_ol_li = edu_array_ol.find('li')
                    if edu_array_ol_li:
                        edu = str(edu_array_ol_li.text.strip())
                        print('Education: ', edu)
                    else:
                        edu = 'Not Available'
                        print('Education: ', edu)
        
                elif edu_array.find('p'):
                    # Handle multiple <p> tags
                    edu_array_ps = edu_array.find_all('p')
                    edu = ''
                    for p in edu_array_ps[:2]:
                        p_text = p.text.strip()
                        if p_text:
                            edu = str(p_text)
                            break 
                    if edu:
                        print('Education: ', edu)
                    else:
                        edu = 'Not Available'
                        print('Education: ', edu)
        
                elif edu_array.find('div'):
                    edu_array_div = edu_array.find('div')
                    edu = str(edu_array_div.text.strip())
                    print('Education (div): ', edu)
        
                else:
                    edu = 'Not Available'
                    print('Education: ', edu)
            else:
                edu = 'Not Available'
                print('Education: ', edu)



            ext_array = soup6.find('ul', class_='teacher__address').findAll('li')[3].text.split()
            if len(ext_array) > 1:
                ext = int(ext_array[1])
                print('Extension: ', ext)
            else:
                ext = 'Not Available'
                print('Extension: ', ext)
        
        faculty_member_dataframe = {
            'ID': id,
            'Name': name,
            'Designation': designation,
            'HEC Approved PHD Supervisor': hec,
            'Highest Education': edu,
            'Email': email,
            'School': school_str,
            'Department': depertment_str,
            'Extension': ext,
            'ImageURL': img
        }
        
        new_dataframe = pd.DataFrame([faculty_member_dataframe])
        faculty_chiniot_data = pd.concat([faculty_chiniot_data, new_dataframe], ignore_index=True)
```

```python
faculty_chiniot_data.to_csv('cfd.csv')
```
