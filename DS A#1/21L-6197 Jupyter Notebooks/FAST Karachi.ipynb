```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
URL_cs = 'https://khi.nu.edu.pk/faculty-php/'
URL_cy = 'https://khi.nu.edu.pk/department-of-cyber-security/'
URL_ai = 'https://khi.nu.edu.pk/department-of-artificial-intelligence/'
URL_se = 'https://khi.nu.edu.pk/department-of-software-engineering/'
URL_ee = 'https://khi.nu.edu.pk/department-of-electrical-engineering/'
URL_ms = 'https://khi.nu.edu.pk/department-of-management-sciences/'
URL_sh = 'https://khi.nu.edu.pk/department-of-sciences-humanities/'

cs = requests.get(URL_cs)
print(cs)
cy = requests.get(URL_cy)
print(cy)
ai = requests.get(URL_ai)
print(ai)
se = requests.get(URL_se)
print(se)
ee = requests.get(URL_ee)
print(ee)
ms = requests.get(URL_ms)
print(ms)
sh = requests.get(URL_sh)
print(sh)
```

```python
soup0 = BeautifulSoup(cs.content, 'html.parser')
soup1 = BeautifulSoup(cy.content, 'html.parser')
soup2 = BeautifulSoup(ai.content, 'html.parser')
soup3 = BeautifulSoup(se.content, 'html.parser')
soup4 = BeautifulSoup(ee.content, 'html.parser')
soup5 = BeautifulSoup(ms.content, 'html.parser')
soup6 = BeautifulSoup(sh.content, 'html.parser')
```

```python
faculty = []
faculty.append(soup0.find('div', class_='gdlr-core-personnel-item').findAll('div', class_='gdlr-core-personnel-list-column'))
faculty.append(soup1.find('div', class_='gdlr-core-personnel-item').findAll('div', class_='gdlr-core-personnel-list-column'))
faculty.append(soup2.find('div', class_='gdlr-core-personnel-item').findAll('div', class_='gdlr-core-personnel-list-column'))
faculty.append(soup3.find('div', class_='gdlr-core-personnel-item').findAll('div', class_='gdlr-core-personnel-list-column'))
faculty.append(soup4.find('div', class_='gdlr-core-personnel-item').findAll('div', class_='gdlr-core-personnel-list-column'))
faculty.append(soup5.find('div', class_='gdlr-core-personnel-item').findAll('div', class_='gdlr-core-personnel-list-column'))
faculty.append(soup6.find('div', class_='gdlr-core-personnel-item').findAll('div', class_='gdlr-core-personnel-list-column'))
```

```python
department_name = []
school_name = []
department_url = 'https://khi.nu.edu.pk/'

department_request = requests.get(department_url)
department_content = BeautifulSoup(department_request.content, 'html.parser')
content = department_content.find('li',class_='menu-item-6256').text.split('\n')
department_name.append(content[4].split('Department of ')[1])
department_name.append(content[5].split('Department of ')[1])
department_name.append(content[6].split('Department of ')[1])
department_name.append(content[7].split('Department of ')[1])
department_name.append(content[10].split('Department of ')[1])
department_name.append(content[11].split('Department of ')[1])
department_name.append(content[12].split('Department of ')[1])

school_name.append(content[2])
school_name.append(content[2])
school_name.append(content[2])
school_name.append(content[2])
school_name.append(content[10])
school_name.append(content[11])
school_name.append(content[12])


print(department_name)
print(school_name)
```

```python
# Create the DataFrame
faculty_karachi_data = pd.DataFrame(columns=['ID',
                                     'Name',
                                     'Designation',
                                     'HEC Approved PHD Supervisor',
                                     'Highest Education',
                                     'Email',
                                     'School',
                                     'Department',
                                     'Extension',
                                     'ImageURL'])
```

```python
faculty_karachi_data.drop(faculty_karachi_data.index, inplace=True)
base_url = 'https://lhr.nu.edu.pk'
count = 0
for faculty_array in faculty:
    depertment = str(department_name[count])
    school =str(school_name[count])
    # print(school)
    count += 1
    for professor in faculty_array:
        print('BIO.............')
        id = 'Not Available'
        print('ID: ',id)
        name  = str(professor.find('div', class_='gdlr-core-personnel-list-content-wrap').find('h3', class_='gdlr-core-personnel-list-title').text.split(',')[0].strip().split('(')[0].strip())
        print('Name: ',name)
        designation = str(professor.find('div', class_='gdlr-core-personnel-list-content-wrap').find('div', class_='gdlr-core-personnel-list-position').text.strip())
        print('Designation: ',designation)
        email = str(professor.find('div', class_='gdlr-core-personnel-list-content-wrap').find('div', class_='gdlr-core-personnel-info').find('div', class_='kingster-personnel-info-list kingster-type-email').text.strip())
        print('Email: ',email)
        print('School: ',school)
        print('Department: ',depertment)
        ext = professor.find('div', class_='gdlr-core-personnel-list-content-wrap').find('div', class_='gdlr-core-personnel-info').find('div', class_='kingster-personnel-info-list kingster-type-phone').text.strip()
        if ext == '-':
            ext = 'Not Available'
            print('Extension: ',ext)
        else:
            print('Extension: ',ext)        
        profile_url = professor.find('div', class_='gdlr-core-personnel-list-content-wrap').find('a', class_='gdlr-core-personnel-list-button').get('href')
        # print('Profile URL: ',profile_url)
        p_url = requests.get(profile_url)
        if p_url.status_code == 200:
            soup7 = BeautifulSoup(p_url.content, 'html.parser')
            hec = soup7.find('span', class_='gdlr-core-title-item-caption').text.split('\n\n')
            # print('HEC: ', hec)
            if len(hec)>1:
                hec = True
                print('HEC: ',hec)
            else:
                hec = False
                print('HEC: ',hec)
                
            # Attempt to find the primary element
            edu_element = soup7.find('span', class_='gdlr-core-icon-list-content')
            
            # Check if the primary element is found and has text
            if edu_element and edu_element.text:
                edu = edu_element.text
                print('Education:', edu)
            else:
                # Attempt to find all elements matching the alternative class
                alt_edu_elements = soup7.findAll('span', class_='gdlr-core-title-item-caption gdlr-core-info-font gdlr-core-skin-caption')
                
                # Check if there are enough elements and the second one has text
                if len(alt_edu_elements) > 1 and alt_edu_elements[1] and alt_edu_elements[1].text:
                    edu = alt_edu_elements[1].text
                    print('Education:', edu)
                else:
                    edu = 'Not available'
                    print('Education:', edu)            

            
        img_url = professor.find('div', class_='gdlr-core-personnel-list-image').find('a').find('img').get('src')
        print('Image URL: ',img_url)
        
        faculty_member_dataframe = {
            'ID': id,
            'Name': name,
            'Designation': designation,
            'HEC Approved PHD Supervisor': hec,
            'Highest Education': edu,
            'Email': email,
            'School': school,
            'Department': depertment,
            'Extension': ext,
            'ImageURL': img_url
        }
        
        new_dataframe = pd.DataFrame([faculty_member_dataframe])
        faculty_karachi_data = pd.concat([faculty_karachi_data, new_dataframe], ignore_index=True)


```

```python
faculty_karachi_data.to_csv('khi.csv')
```
